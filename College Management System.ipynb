```python
import mysql.connector
import getpass
```

```python
mydb=mysql.connector.connect(
  host="127.0.0.1",
  user="root", 
  database="Collegedata"
)
print(mydb)#to verify
```

```python
mycursor=mydb.cursor()
mycursor.execute("CREATE TABLE userdata(ID INT NOT NULL AUTO_INCREMENT PRIMARY KEY,NAME VARCHAR(255),PASSWORD VARCHAR(255),PRIVILEGE VARCHAR(50))")
```

```python
mycursor=mydb.cursor()
admin="Vaibhaav"
adminpassword="1to8"
def admin_login():
    while 1:
        print()
        print("Admin menu")
        print("1. Register new student")
        print("2. Register new teacher")
        print("3. Delete Existing student")
        print("4. Delete Existing teacher")
        print("5. Exit")
        user_input=input("Option: ")
        if user_input=='1':
            name=input("Student username: ")
            password=input("Student password: ")
            values=(name,password)
            mycursor.execute("INSERT INTO userdata (NAME,PASSWORD,PRIVILEGE) VALUES(%s,%s,'student')",values)
            mydb.commit()
            print(name+" has been inserted as student")
        elif user_input == '2':
            name=input("Teacher username: ")
            password=input("Teacher password: ")
            values=(name,password)
            mycursor.execute("INSERT INTO userdata (NAME,PASSWORD,PRIVILEGE) VALUES(%s,%s,'teacher')",values)
            mydb.commit()
            print(name+" has been inserted as teacher")
        elif user_input == '3':
            print()
            print("Delete Existing Student account \n")
            name=input("Student username: ")
            value=(name,'student')
            mycursor.execute("DELETE FROM userdata WHERE NAME= %s AND PRIVILEGE=%s",value)
            mydb.commit()    
            if mycursor.rowcount < 1:
                print("no data exist \n")
            else:
                print(name + " deleted successfully \n")
            
        elif user_input == '4':
            print()
            print("Delete Existing Teacher account \n")
            name=input("Teacher username: ")
            value=(name,'teacher')
            mycursor.execute("DELETE FROM userdata WHERE NAME= %s AND PRIVILEGE=%s",value)
            mydb.commit()
            if mycursor.rowcount < 1:
                print("no data exist \n")
            else:
                print(name + " deleted successfully \n")
                
            
        elif user_input == '5':
            print("Logging out")
            break
        else:
            print("Invalid input \n")
            
            
            
def verify_admin():
    print()
    print("Admin login")
    print()
    username=input("Username: ")
    passwrd=getpass.getpass("password: ")
    if username== admin:
        if passwrd==adminpassword:
            print("Verification successful \n")
            admin_login()
    else:
        print("Wrong Creditionals \n")
def main():
    while 1:
        print()
        print("Welcome to the college management system")
        print("Press 1 to login as admin or -1 to exit from system")
        user_option =input()
        if user_option =='1':
            verify_admin()
        elif user_option == '-1':
            print("Exiting system..... \n")
            print("Have a great day ahead :)")
            break
        else:
            print("invalid input")
main()
```
